```python
import numpy as np
import torch
from pytorch_metric_learning.losses import TripletMarginLoss
from scipy import stats
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
from pytorch_metric_learning.distances import LpDistance
```

```python
def normal_loss(c, scale):
    criterion = TripletMarginLoss(triplets_per_anchor=1000, distance=LpDistance(normalize_embeddings=False, p=2))
    l_rv = stats.norm(loc=c, scale=scale)
    r_rv = stats.norm(scale=scale)
    l = l_rv.rvs(100).reshape((50, 2))
    r = r_rv.rvs(100).reshape((50, 2))
    df = pd.DataFrame()
    df['x'] = np.concatenate((l[:, 0], r[:, 0]))
    df['y'] = np.concatenate((l[:, 1], r[:, 1]))
    df['label'] = np.concatenate((np.zeros(50), np.ones(50)))
    #fig = px.scatter(df, x='x', y='y', color='label')
    #fig.show()
    embeddings = torch.as_tensor(np.concatenate((l, r)))
    labels = torch.as_tensor(df['label'])
    loss = criterion(embeddings, labels)
    print(f'center = {c}, loss = {loss}')
    return loss
```

```python
def test_triplet_loss():
    
    for c in np.arange(-1., 0.1, 0.1):
        normal_loss(c)
```
